```python
import os
from google.colab import userdata

os.environ["GROQ_API_KEY"] = userdata.get("GROQ_API_KEY")
```

```python
from groq import Groq

client=Groq(api_key=os.environ["GROQ_API_KEY"])
```

```python
import gradio as gr


def respond(message, history, System_promt, Temperature):
  if history is None:
    history = []

  messages = []
  if System_promt:
    messages.append({"role": "system", "content": System_promt})
  for turn in history:
        messages.append({"role": turn['role'], "content": turn['content']})
  messages.append({"role": "user", "content": message})
  stream = client.chat.completions.create(
    model="llama-3.1-8b-instant",
    messages=messages,
    temperature=Temperature,
    stream=True,

  )
  partial=""
  for chunk in stream:
    partial+=chunk.choices[0].delta.content or ""
    yield partial


demo = gr.ChatInterface(
    fn=respond,
    chatbot=gr.Chatbot(height=300, allow_tags=True),
    textbox=gr.Textbox(placeholder="Ask me anything", container=False, scale=7),
    title="AI Expert",
    theme="soft",
    additional_inputs=[
        gr.Textbox(label="System Prompt", value="", lines=3),
        gr.Slider(minimum=0.0, maximum=1.0, step=0.1, value=0.7, label="Temperature")
    ],
    type="messages"
)
demo.launch(debug=True)
```
